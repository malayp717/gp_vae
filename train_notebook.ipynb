```python
from pathlib import Path
from types import SimpleNamespace
import sys

ROOT = Path.cwd()
SRC = ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

from vae.cli.app import build_overrides, setup_logging
from vae.training.engine import train

setup_logging(verbose)

train_args = SimpleNamespace(model=model, lpips=lpips, adv=adv)
config_overrides = build_overrides(train_args)
config_overrides
```
